NLP Natural Language Processing

```python
import nltk
# nltk.download("all")
from nltk.corpus import stopwords
from string import punctuation
```

```python
p1 = "hello world. I am here to help you. I forgot your name! can you tell me again."
```

```python
token = nltk.sent_tokenize(p1)

token
```

```python
token2 = nltk.word_tokenize(p1)

token2
```

```python
p3 = nltk.pos_tag(token2)
p3
```

```python
pt = list(punctuation)
pt
```

```python
sw = stopwords.words("English")
sw
```

```python
swpt = sw + pt

swpt
```

```python
para="I am a boy. Hi! how are you? I am fine well and how about you?"
para2 = nltk.word_tokenize(para)
para2
```

```python
for i in para2:
    if i not in swpt:
        print(i)
```
