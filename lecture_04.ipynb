```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
plt.xlabel("area[sq feet]")
plt.ylabel("price")
plt.scatter(df.area,df.price,color="red",marker="+")
plt
```
