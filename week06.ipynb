cost function in linear regressions:

```python
import numpy as np
import matplotlib.pyplot as plt

# Sample data (house size in square feet and price in $1000s)
X = np.array([600, 800, 1000, 1200, 1400])  # House sizes (input feature)
y = np.array([200, 250, 300, 350, 400])     # Prices (target values)
```

```python
def hypothesis(theta, X):
    return theta[0] + theta[1] * X 
```

```python
def cost_function(theta, X, y):
    m = len(y)  # Number of training examples
    predictions = hypothesis(theta, X)
    errors = predictions - y
    return (1/(2*m)) * np.sum(errors ** 2)
```

```python
theta_test_1 = np.array([0, 0])  # Bad guess
theta_test_2 = np.array([50, 0.25])  # Closer guess
theta_test_3 = np.array([100, 0.2])  # Even closer

# Compute cost for each guess
cost1 = cost_function(theta_test_1, X, y)
cost2 = cost_function(theta_test_2, X, y)
cost3 = cost_function(theta_test_3, X, y)

print(f"Cost for theta = [0, 0]: {cost1}")
print(f"Cost for theta = [50, 0.25]: {cost2}")
print(f"Cost for theta = [100, 0.2]: {cost3}")
```

```python
# Plot the data points and the different hypothesis functions
plt.scatter(X, y, color='red', marker='x', label='Training data')

# Plot for theta_test_1
plt.plot(X, hypothesis(theta_test_1, X), label='Hypothesis: theta = [0, 0]')
# Plot for theta_test_2
plt.plot(X, hypothesis(theta_test_2, X), label='Hypothesis: theta = [50, 0.25]')
# Plot for theta_test_3
plt.plot(X, hypothesis(theta_test_3, X), label='Hypothesis: theta = [100, 0.2]')

plt.xlabel('Size of house (sq.ft)')
plt.ylabel('Price ($1000s)')
plt.title('House Price vs. Size')
plt.legend()
plt.show()
```

gradient descent in linear regression

```python
import numpy as np
import matplotlib.pyplot as plt

```

```python

# Define the cost function
def cost_function(x):
    return x ** 2
```

```python

# Define the gradient of the cost function
def gradient(x):
    return 2 * x
```

```python

# Gradient Descent function
def gradient_descent(starting_x, learning_rate, num_iterations):
    x = starting_x
    x_history = [x]  # Keep track of x values for visualization
    cost_history = [cost_function(x)]  # Keep track of cost values
    
    for i in range(num_iterations):
        grad = gradient(x)
        x = x - learning_rate * grad
        x_history.append(x)
        cost_history.append(cost_function(x))
    
    return x, x_history, cost_history
```

```python

# Parameters
starting_x = 10  # Initial guess
learning_rate = 0.1  # Step size
num_iterations = 100  # Number of iterations
```

```python

# Run gradient descent
optimal_x, x_history, cost_history = gradient_descent(starting_x, learning_rate, num_iterations)

# Print final result
print(f'Optimal x: {optimal_x}')
print(f'Cost at optimal x: {cost_function(optimal_x)}')
```

```python

# Plot cost function history
plt.plot(cost_history)
plt.title('Cost Function over Iterations')
plt.xlabel('Iteration')
plt.ylabel('Cost')
plt.show()

# Plot x values over iterations
plt.plot(x_history)
plt.title('X values over Iterations')
plt.xlabel('Iteration')
plt.ylabel('x value')
plt.show()
```

multivariate regression

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Example data: house size (in sq feet), number of bedrooms, price
X = np.array([[1500, 3], [1600, 3], [1700, 4], [1800, 3], [1900, 4]])  # Size and Bedrooms
y = np.array([300000, 320000, 350000, 370000, 400000])  # Price

# Fit the model
model = LinearRegression()
model.fit(X, y)

# Predict price for a new house
new_house = np.array([[1650, 3]])
predicted_price = model.predict(new_house)
print(f"Predicted price: {predicted_price[0]}")
```

```python
plt.plot(X,y)
plt.title('house predictions')
plt.xlabel('size & bedrooms')
plt.ylabel('price')
plt.show()
```

Polynomial regression

```python
import numpy as np
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression

# Example data: Speed vs Stopping Distance
X = np.array([[10], [20], [30], [40], [50]])  # Speed
y = np.array([9, 36, 81, 144, 225])  # Stopping distance

# Transform the features into polynomial features (degree 2)
poly = PolynomialFeatures(degree=2)
X_poly = poly.fit_transform(X)

# Fit the model
model = LinearRegression()
model.fit(X_poly, y)

# Predict stopping distance for speed 60 
predicted_distance = model.predict(poly.transform([[60]]))
print(f"Predicted stopping distance: {predicted_distance[0]}")
```

#Logistic regression

```python
import numpy as np
from sklearn.linear_model import LogisticRegression

# Example data: Age, Income, Buy (1/0)
X = np.array([[25, 50000], [35, 60000], [45, 80000], [30, 54000], [40, 75000]])  # Age and Income
y = np.array([0, 0, 1, 0, 1])  # Buy or not

# Fit the model
model = LogisticRegression()
model.fit(X, y)

# Predict whether a new customer (age 28, income 58k) will buy
new_customer = np.array([[28, 68000]])
predicted_class = model.predict(new_customer)
predicted_probability = model.predict_proba(new_customer)
print(f"Predicted class: {predicted_class[0]}, Probability: {predicted_probability[0][1]}")
```
