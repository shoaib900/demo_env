cost function in linear regressions:

```python
import numpy as np
import matplotlib.pyplot as plt

# Sample data (house size in square feet and price in $1000s)
X = np.array([600, 800, 1000, 1200, 1400])  # House sizes (input feature)
y = np.array([200, 250, 300, 350, 400])     # Prices (target values)
```

```python
def hypothesis(theta, X):
    return theta[0] + theta[1] * X 
```

```python
def cost_function(theta, X, y):
    m = len(y)  # Number of training examples
    predictions = hypothesis(theta, X)
    errors = predictions - y
    return (1/(2*m)) * np.sum(errors ** 2)
```

```python
theta_test_1 = np.array([0, 0])  # Bad guess
theta_test_2 = np.array([50, 0.25])  # Closer guess
theta_test_3 = np.array([100, 0.2])  # Even closer

# Compute cost for each guess
cost1 = cost_function(theta_test_1, X, y)
cost2 = cost_function(theta_test_2, X, y)
cost3 = cost_function(theta_test_3, X, y)

print(f"Cost for theta = [0, 0]: {cost1}")
print(f"Cost for theta = [50, 0.25]: {cost2}")
print(f"Cost for theta = [100, 0.2]: {cost3}")
```

```python
# Plot the data points and the different hypothesis functions
plt.scatter(X, y, color='red', marker='x', label='Training data')

# Plot for theta_test_1
plt.plot(X, hypothesis(theta_test_1, X), label='Hypothesis: theta = [0, 0]')
# Plot for theta_test_2
plt.plot(X, hypothesis(theta_test_2, X), label='Hypothesis: theta = [50, 0.25]')
# Plot for theta_test_3
plt.plot(X, hypothesis(theta_test_3, X), label='Hypothesis: theta = [100, 0.2]')

plt.xlabel('Size of house (sq.ft)')
plt.ylabel('Price ($1000s)')
plt.title('House Price vs. Size')
plt.legend()
plt.show()
```
